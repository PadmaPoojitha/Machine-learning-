```python
import pandas as pd
import numpy as np
import keras 
from keras.utils.vis_utils import plot_model
import keras.models as km
import keras.layers as kl
```

```python
training_data = pd.read_csv("/train.csv").to_numpy() 
testing_data = pd.read_csv("/test.csv").to_numpy()
model = km.Sequential()
model.add(kl.Dense(32, input_dim = 7, use_bias = True, bias_initializer = "ones", activation = "tanh"))
model.add(kl.Dense(4, use_bias = True, bias_initializer = "ones", activation = "tanh"))
model.add(kl.Dense(1, use_bias = True, bias_initializer = "ones", activation = "tanh"))
model.compile(loss="binary_crossentropy", optimizer = "rmsprop", metrics=["accuracy"])
plot_model(model, to_file='plot_1.png')
```

```python
traindata_x = training_data[:,0:7] 
traindata_y = training_data[:,7]
model.fit(traindata_x, traindata_y, epochs=1000) 
model.evaluate(traindata_x, traindata_y)
testdata_x = testing_data[:,0:7] 
testdata_y = testing_data[:,7]
prediction = model.predict(testdata_x) 
print(prediction)
```
